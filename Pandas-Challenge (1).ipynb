# Purchasing analysis tool 

```python
# identifying unique items
item_count = len(purchase_data_df["Item Name"].unique())
ave_price = purchase_data_df["Price"].mean()
purchase_num = purchase_data_df["Purchase ID"].count()
total_rev = purchase_data_df["Price"].sum()
summary_table = pd.DataFrame({"Numbe of unique items": [item_count], #why do i need brackets
                             "Average Price": ave_price, 
                             "Number of Purchases": purchase_num, 
                             "Total Revenue": total_rev})
summary_table
```

# Gender Demographics

```python
 #Gender Demographics
male_count = purchase_data_df[purchase_data_df.Gender == "Male"]
male_count_int = len(male_count["SN"].unique())
female_count = purchase_data_df[purchase_data_df.Gender == "Female"]
female_count_int = len(female_count["SN"].unique())
#other count
non_disc = purchase_data_df[purchase_data_df.Gender == "Other / Non-Disclosed"]
non_disc_count_int = len(non_disc["SN"].unique())

#gender total
gender_total = male_count_int + female_count_int


#male percent
male_percentage = male_count_int/gender_total
#female percent
female_percentage = female_count_int/gender_total
#other percent
non_disc_percent = non_disc_count_int/gender_total


gender_summary_table = pd.DataFrame({"Total Players": [male_count_int, female_count_int, non_disc_count_int],
                                     "Percentage of Players": [male_percentage, female_percentage, non_disc_percent]}, index = ["Male", "Female", "Other/Non-Disclosed"])
gender_summary_table
```

# Purchasing analysis (gender) 

```python


#gender purchase count
male_purch_count = len(male_count["Purchase ID"].unique())
female_purch_count = len(female_count["Purchase ID"].unique())
other_purch_count = len(non_disc["Purchase ID"].unique())

#gender purchase percentage
#male average price
ave_price_male = purchase_data_df[purchase_data_df["Gender"] == "Male"]
ave_price_male = ave_price_male["Price"].describe()
ave_price_male[1]

#Female average price
ave_price_female = purchase_data_df[purchase_data_df["Gender"] == "Female"]
ave_price_female = ave_price_female["Price"].describe()
ave_price_female[1]

#Other average price
ave_price_other = purchase_data_df[purchase_data_df["Gender"] == "Other / Non-Disclosed"]
ave_price_other = ave_price_other["Price"].describe()
```

```python
#Total purchase value

#male purchase value
total_val_male = purchase_data_df[purchase_data_df["Gender"] == "Male"]
total_val_male = total_val_male["Price"].sum()
#female purchase value
total_val_female = purchase_data_df[purchase_data_df["Gender"] == "Female"]
total_val_female = total_val_female["Price"].sum()

#other purchase value
total_val_other = purchase_data_df[purchase_data_df["Gender"] == "Other / Non-Disclosed"]
total_val_other = total_val_other["Price"].sum()

#average total purchase per person

#male
ave_total_purch_m = total_val_male / male_count_int
#female
ave_total_purch_f = total_val_female / female_count_int
#other
ave_total_purch_o = total_val_other / non_disc_count_int

#data frame summary 
gender_purch_sum = pd.DataFrame({"Purchase Count": [male_purch_count, female_purch_count, ave_price_other[0]],
                                 "Average Purchase Price": [ave_price_male[1], ave_price_female[1], ave_price_other[1]],
                                 "Total Value Purchase": [total_val_male, total_val_female, total_val_other],
                                 "Average Total Purchase Per Person": [ave_total_purch_m, ave_total_purch_f, ave_total_purch_o]},
                                index = ["Male", "Female", "Other /Non-Disclosed"])

gender_purch_sum
```

# Age Demographics

```python
# Create age groups
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 4000]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
df_age_ranges = pd.DataFrame({
    "Age Ranges": pd.cut(purchase_data_df["Age"], bins=bins, labels=labels),
})

# Calculate count of each group
age_range_count = df_age_ranges["Age Ranges"].value_counts()
total_players = purchase_data_df.count()[0]
df_age_demos = pd.DataFrame({
    "Total Count": age_range_count,
    "Percentage of Players": round(age_range_count / total_players * 100, 2)
})


df_age_demos.sort_index()
```

# Purchasing Analysis (Age)

```python
purchase_data_df_copy = purchase_data_df.copy()
purchase_data_df_copy = purchase_data_df_copy.loc[:, ["Price"]]
purchase_data_df_copy["Age Ranges"] = pd.cut(purchase_data_df["Age"], bins=bins, labels=labels)

# Calculate values by Age Range
purchase_mean = purchase_data_df_copy.groupby(["Age Ranges"]).mean()["Price"]
# TODO: Sum purchase price
purchase_sum = purchase_data_df_copy.groupby(["Age Ranges"]).sum()["Price"]
# TODO: Avg Total per person
ave_total_purch_person = purchase_sum/age_range_count

# Throw the calculated values into a Data Frame
df_purchase_analysis = pd.DataFrame({
    "Average Purchase Price": purchase_mean,
    "Total Purchase Value": purchase_sum,
    "Average Total Purchases per Person": round(ave_total_purch_person * 1,2)})

# Format the new Data Frame for readability
df_purchase_analysis = df_purchase_analysis.style.format({
    "Average Purchase Price": "${:,.2f}",
    "Average Total Purchases per Person": "${:,.2f}"
})

df_purchase_analysis
```

# Top spenders


```python

pd_copy = purchase_data_df.copy()
#settign index to "SN"
pd_copy.set_index("SN")

#creating columns
sn_purch_count = pd_copy.groupby(["SN"]).count()["Purchase ID"]
sn_ave_purch_price = pd_copy.groupby("SN").mean()["Price"]
sn_total_purch_val = pd_copy.groupby("SN").sum()["Price"]

#create summary data frame
pd_sum = pd.DataFrame({ "Purchase Count": sn_purch_count,
                            "Average Purchase Price": sn_ave_purch_price, 
                            "Total Purch Value": sn_total_purch_val})


#ascending 'Total Purch Value'
pd_sum_sorted = pd_sum.sort_values("Total Purch Value", ascending = False)
pd_sum_sorted
```
